```python
import numpy as np
import cv2 as cv

path = r'/content/drive/MyDrive/Colab Notebooks/Untitled.png'
im = cv.imread(path)
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

sortedArea = sorted(contours, key=cv.contourArea, reverse=True)


cv2.drawContours(im, sortedArea, 1, (0, 0, 255), 3)
from google.colab.patches import cv2_imshow
cv2_imshow(im)
cv2.waitKey(0)
```
